```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x));
ax.set_title('simple plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label');
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.plot(x,np.cos(x));
ax.set_title('multiple lin');
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.legend()
#ax.legend(loc=1)
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.plot(x,np.sin(x),label='sin',color='red')
ax.plot(x,np.cos(x),label='cos',color='blue')
ax.set_title('multiple line');
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.legend();
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos+1',linestyle='dotted')
ax.plot(x,np.sin(x+1),label='cos',linestyle='dashdot')
ax.set_title('multiple lines');
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend();
#ax.legend(loc=1)
plt.show()
```
